```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
import numpy as np
```

```python
df=pd.read_csv('amazon_laptop_Data.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df[['asin','title','link_clean','thumbnail','rating','reviews','extracted_price','specs','delivery']]
```

```python

def handle_missing_values(df):
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    
    transformer = ColumnTransformer(
        transformers=[
            ('num_imputer', SimpleImputer(strategy='mean'), numerical_cols),
            ('cat_imputer', SimpleImputer(strategy='most_frequent'), categorical_cols)
        ]
    )
    transformed_data = transformer.fit_transform(df)
    
    all_cols = numerical_cols + categorical_cols
    df_imputed = pd.DataFrame(transformed_data, columns=all_cols, index=df.index)
    
    return df_imputed
```

```python
df=handle_missing_values(df)
```

```python
def clean_delivery(col):
    print(col[2:15])
    return col[2:15]
    
        
```

```python
df['delivery']=df['delivery'].apply(clean_delivery)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
def datatype_conversion(df,col):
    return df[col].astype(float)
```

```python
def dollar_to_ruppe(col):
    return int(col*86.46)
```

```python
df[['reviews','rating','extracted_price']]=datatype_conversion(df,['reviews','rating','extracted_price'])
```

```python
df['Price(indian_rupees)']=df['extracted_price'].apply(dollar_to_ruppe)
```

```python
df
```

```python
import ast

def parse_specs(spec_str):
    try:
        spec_dict = ast.literal_eval(spec_str)
        return pd.Series({
            'display_size': spec_dict.get('display_size'),
            'ram': spec_dict.get('ram'),
            'disk_size': spec_dict.get('disk_size'),
            'operating_system': spec_dict.get('operating_system')
        })
    except:
        return pd.Series({
            'display_size': None,
            'ram': None,
            'disk_size': None,
            'operating_system': None
        })

```

```python
specs_df = df['specs'].apply(parse_specs)
df = pd.concat([df, specs_df], axis=1)
df.drop(columns=['specs'], inplace=True)
```
