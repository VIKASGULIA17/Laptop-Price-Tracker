```python
laptops = []

for i in response.get("organic_results", []):
    laptop = {
        "asin":i.get("asin"),
        "title": i.get("title"),
        "thumbnail": i.get("thumbnail"),
        "rating": i.get("rating"),
        "price": i.get("extracted_price"),
        "old_price":i.get("extracted_old_price"),
        "link": i.get("link_clean"),
        "specs": i.get("specs")
    }
    laptops.append(laptop)
```

```python
page=2
all_other_pages=response.get("organic_results",[])
while True:
    params['page']=page
    search=requests.get("https://serpapi.com/search",params=params)
    response=search.json()

    if "organic_results"not in response or not response['organic_results']:
        break

    organic_results=response['organic_results']
    all_other_pages.extend(organic_results)
    print(f"Page {page} processed.\n")
    page+=1
    
```
